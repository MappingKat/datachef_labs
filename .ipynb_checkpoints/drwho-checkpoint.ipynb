# Dr Who Target Visualization Code

```python
import pandas as pd
import re
from ggplot import *
%matplotlib inline
import numpy as np
import string
```

```python
df = pd.read_csv('https://www.dropbox.com/s/ieedjb59qhs30ma/drwho.csv?dl=1')
```

```python
df.columns
```

```python
df.head()
```

```python
df.index = df['Villain']
```

```python
df = df.sort('Epi- sodes', ascending=False)
```

## Difficulty: Easy

```python
# Top 25 Most Frequently Seen
df['Epi- sodes'][0:25].plot(kind='bar')
```

## Difficulty: Harder

```python
df['duration'] = df['Year last'] - df['Year first']
```

```python
df = df.sort('duration', ascending=False)
```

```python
# Top 25 Longest Running Dr. Who Villans
df['duration'][0:25].plot(kind='bar')
```
