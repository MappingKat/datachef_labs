# Game Reviews Target Visualization Code

```python
import pandas as pd
import re
from ggplot import *
%matplotlib inline
import numpy as np
import string
```

```python
df = pd.read_csv('https://www.dropbox.com/s/ehcamae0ivjvkux/gamedata.csv?dl=1')
```

```python
df.columns
```

```python
df.head()
```

```python
df_by_scores = df.groupby('Platform').mean()
```

```python
df_by_scores = df_by_scores.sort('Score', ascending=False)
```

```python
df_by_scores
```

```python
# Top 10 Mean Game Scores By Platform
df_by_scores[0:10].plot(kind='bar')
```

```python
df_by_scores = df_by_scores.sort('Score', ascending=True)
```

```python
# Bottom 10 Mean Game Scores By Platform
df_by_scores[0:10].plot(kind='bar')
```
