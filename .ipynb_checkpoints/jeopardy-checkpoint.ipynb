# Jeopardy Questions Target Visualization Code

```python
import pandas as pd
import re
from ggplot import *
%matplotlib inline
import numpy as np
import string
```

```python
df = pd.read_csv('https://www.dropbox.com/s/0sxhkjzyczjymhl/JEOPARDY_CSV.csv?dl=1')
```

```python
df.columns
```

```python
replacer = lambda x: re.sub('[$,None]', '', x)
```

```python
df[' Value'] = df[' Value'].map(replacer)
```

```python
df[' Value'] = df[' Value'].replace('', np.nan)
```

```python
df[' Value'] = df[' Value'].fillna(0)
```

```python
df[' Value'] = df[' Value'].astype(int)
```

```python
df_by_cat = df.groupby(df[' Category']).sum()
```

```python
df_by_cat = df_by_cat.sort(columns=' Value', ascending=False)
```

```python
# Which 25 Jeopardy Categories Have Been Worth The Most Money?
df_by_cat[' Value'][0:25].plot(kind='bar')
```
