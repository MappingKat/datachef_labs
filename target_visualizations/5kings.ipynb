# Five Kings Target Visualization Code

```python
import pandas as pd
from ggplot import *
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('https://www.dropbox.com/sh/7a9r3gv93yctjc0/AADhbzYj1jxzRc2pc2QO1DmLa/5kings_battles_v1.csv?dl=1')
```

## Y: Defending Army Size, X Attacker Army Size, Color of markers: battle outcome

```python
ggplot(aes(x='attacker_size', y='defender_size', colour='attacker_outcome'), 
       data=df[df['attacker_size'].notnull() & df['defender_size'].notnull() & df['attacker_outcome'].notnull()]) + \
    geom_point()
```
