## Load the test dataset

```python
# Load the test dataset
data_df = pd.read_csv('test_data.csv', encoding='ISO-8859-1')
```

## Load the dataset with the true variable types

```python
# Load the dataset containing the true variables
var_dataset = pd.read_csv('var_type_test.csv')
```

```python
# View the top few rows
var_dataset.head(3)
```

## Clean up the missing values

```python
# Create a list of values that denote missing values
missing_values = [-1111, -999, '.', 'None', 'none', 'x', '__NA__']

# Replace everything in the list above with np.nan
data_df = data_df.replace(missing_values, np.nan)

data_df = data_df.convert_objects()
```

## Create the classify function

```python
# Create a list called type_guess, which is where 
# the classifer function will place it's guesses.
type_guess = []

# For each column in in_df,
for c in data_df.columns:
    # If a particular in_df column is all nan values,
    if len(data_df[c].unique()) == 1 and str(data_df[c][0]) == 'nan':
        type_guess.append('empty')
    else:
        if 'latitude'in c.lower():
            type_guess.append('geo')
        elif 'longitude'in c.lower():
            type_guess.append('geo')
        elif 'zip'in c.lower():
            type_guess.append('geo')
        elif 'street'in c.lower():
            type_guess.append('geo')
        elif 'city'in c.lower():
            type_guess.append('geo')
        elif 'state'in c.lower():
            type_guess.append('geo')
        elif 'address'in c.lower():
            type_guess.append('geo')
        elif 'geo_lat'in c.lower():
            type_guess.append('geo')
        elif 'geo_lon'in c.lower():
            type_guess.append('geo')
        elif 'postcode'in c.lower():
            type_guess.append('geo')
        elif 'fips'in c.lower():
            type_guess.append('geo')
        elif 'county'in c.lower():
            type_guess.append('geo')
        elif 'country'in c.lower():
            type_guess.append('geo')
        elif 'date'in c.lower():
            type_guess.append('datetime')
        elif 'year'in c.lower():
            type_guess.append('datetime')
        elif 'time'in c.lower():
            type_guess.append('datetime')
        elif 'id'in c.lower():
            type_guess.append('nominal')
        elif type(data_df[c][data_df[c].first_valid_index()]) == np.float64:
            if len(data_df[c].unique()) < 10:
                type_guess.append('discrete')
            else:
                type_guess.append('continuous')
        elif type(data_df[c][data_df[c].first_valid_index()]) == np.int64:
            if len(str(data_df[c][data_df[c].first_valid_index()])) >= 2:
                type_guess.append('discrete')
            elif len(data_df[c].unique()) < 10:
                type_guess.append('nominal')
            else:
                type_guess.append('discrete')
        elif type(data_df[c][data_df[c].first_valid_index()]) == np.float64:
            type_guess.append('continuous')
        elif type(data_df[c][data_df[c].first_valid_index()]) == str:
            if len(data_df[c].unique()) < 10:
                type_guess.append('nominal')
            else:
                type_guess.append('unstructured')
        else:
            type_guess.append('unknown')

var_dataset['type_guess'] = type_guess
```

## Tabulate the function's accuracy against the true variable types

```python
var_dataset
```

```python
var_dataset['type_score'] = np.where(var_dataset['type_guess']==var_dataset['type_true'], True, False)
```

```python
print('SCORE CARD:')

# Overall accuracy percent:
print('Overall Accuracy:', var_dataset['type_score'].value_counts()[1]/len(var_dataset['type_score'])*100)

# Accuracy of nominal variables
print('Nominal Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'nominal'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of continuous variables
print('Continuous Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'continuous'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of unstructured variables
print('Unstructured Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'unstructured'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of empty variables
print('Discrete Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'discrete'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of empty variables
print('Empty Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'empty'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of geo variables
print('Geo Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'geo'].value_counts(normalize=True, sort=False)[-1]*100)

# Accuracy of datetime variables
print('Datetime Accuracy:', var_dataset['type_score'][var_dataset['type_true'] == 'datetime'].value_counts(normalize=True, sort=False)[-1]*100)
```
